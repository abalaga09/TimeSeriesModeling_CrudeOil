You can see how much memory you have available at any time by running the following code cell.
```
#12 years worth of historical data for WTI Crude Oil
import numpy as np
import pandas as pd
import yfinance as yf

tickers = ['CL=F']

ticker_price = yf.download(tickers, start='2024-01-01')


ticker_price
```

```
closing_price = ticker_price['Close']

#convert close price to returns
ret = closing_price.pct_change()

ret['Volume'] = ticker_price['Volume']
ret.columns = ['Returns', 'Volume']
ret
```

```
ret = ret.reset_index()  # Moves Date index into a column
ret['Date'] = pd.to_datetime(ret['Date'])  # Now this will work
```

```
#oil_csv = ret.to_csv('crude_oil.csv')
ret.columns
```

```
#implement time series data analysis
ret['Returns'].plot()
ret_cleaned = ret['Returns'].dropna()
ret_cleaned
```

```
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
```

```
decomp = seasonal_decompose(ret_cleaned, model='product',period=30)
decomp.plot()
plt.show()
```

```
from statsmodels.tsa.seasonal import STL
stl = STL(ret_cleaned, period= 12)
res = stl.fit()
res.plot()
plt.suptitle('STL Decomposition')
plt.tight_layout()
plt.show()
#extract seasonal trend data from 6 years of historical data and then analyze (use close prices)
```

```
#started to look for stationary or non-stationary
from statsmodels.tsa.stattools import adfuller
stationary = adfuller(ret_cleaned)
print(f"Augmented-Dickey-Fuller Test Statistic:{stationary[0]}")
print(f"P-Value:{stationary[1]}")
```

```
#find parameters
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```
plot_acf(ret_cleaned)
plot_pacf(ret_cleaned)
plt.show()
```

```
#implementing SARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```
model = SARIMAX(ret_cleaned, order=(2,0,2), seasonal_order=(1,1,1,12))
result = model.fit()
print(result.summary())
```

```
#testing and training part to our dataset/forecasting
train = ret_cleaned[:-365]
test = ret_cleaned[-365:]

forecast = result.forecast(steps=365)
```
